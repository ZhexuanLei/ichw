```python
"""planets.py : Description of the revolving around the sun of the 6 Planets
_author_ = "雷哲轩"
_pkuid_ = "1700011714"
_pkuemail_ = "lzx1700011714@pku.edu.cn"
"""

import turtle
import math
pm = turtle.Screen()
pm.screensize(500,500)
pm.bgcolor('black')
mer = turtle.Turtle()
ven = turtle.Turtle()
ear = turtle.Turtle()
mar = turtle.Turtle()
jup = turtle.Turtle()
sat = turtle.Turtle()
Sun = turtle.Turtle()
mer.shape('circle')
ven.shape('circle')
ear.shape('circle')
mar.shape('circle')
jup.shape('circle')
sat.shape('circle')
Sun.shape('circle')
mer.shapesize(0.1)
ven.shapesize(0.15)
ear.shapesize(0.2)
mar.shapesize(0.25)
jup.shapesize(0.3)
sat.shapesize(0.35)
Sun.shapesize(0.5)
mer.color('white')
ven.color('green')
ear.color('blue')
mar.color('orange')
jup.color('purple')
sat.color('grey')
Sun.color('red')
mer.pu()
ven.pu()
ear.pu()
mar.pu()
jup.pu()
sat.pu()
mer.fd(33)
ven.fd(104)
ear.fd(171)
mar.fd(220)
jup.fd(280)
sat.fd(331)
mer.pd()
ven.pd()
ear.pd()
mar.pd()
jup.pd()
sat.pd()
mer.left(90)
ven.left(90)
ear.left(90)
mar.left(90)
jup.left(90)
sat.left(90)
mer.pensize(0.5)
ven.pensize(0.75)
ear.pensize(1)
mar.pensize(1.25)
jup.pensize(1.5)
sat.pensize(1.35)
for i in range(1,1200,1):
    x1=20*math.cos(math.radians(7.85*i))+13
    y1=15*math.sin(math.radians(7.85*i))
    mer.goto(x1,y1)
    x2=60*math.cos(math.radians(3.08*i))+44
    y2=40*math.sin(math.radians(3.08*i))
    ven.goto(x2,y2)
    x3=100*math.cos(math.radians(1.88*i))+71
    y3=70*math.sin(math.radians(1.88*i))
    ear.goto(x3,y3)
    x4=140*math.cos(math.radians(i))+80
    y4=115*math.sin(math.radians(i))
    mar.goto(x4,y4)               
    x5=180*math.cos(math.radians(0.5*i))+100
    y5=150*math.sin(math.radians(0.5*i))
    jup.goto(x5,y5)
    x6=220*math.cos(math.radians(0.3*i))+111
    y6=190*math.sin(math.radians(0.3*i))
    sat.goto(x6,y6)
turtle.exitonclick  
```
